```python
# Import the required libraries:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Check the dataset:

df=pd.read_csv("TedTalks.csv")

df.head(2)
```

```python
# Check the D-type of variables then will convert some.

df.info() #change the D-type of some of columns
```

```python
# Check the missing values:

missing_values = round(df.isnull().sum()/len(df),2)*100
missing_values
```

```python
# Convert the D-type of the data-frame

df["Event"] = df["Event"].astype("string")
df["Published_Date"] = pd.to_datetime(df["Published_Date"])
df["Title"] = df["Title"].astype("string")
```

```python
# Remove the letter "k" from the Likes column

df["Likes"] = df["Likes"].str.replace("K","").str.replace("M","")
```

```python
# Now our variable is ready to convert

df["Likes"] = df["Likes"].astype(float)
```

```python
# Double check that everything works fine..

df.info() 
```

#### We still need to work on those three variables ["Speakers", "Subtitles", "Topics"]

```python
# Check the statistical values for each column:

round(df.describe().T,2)
```

```python
data = df['Speakers'].str.split(',', 1, expand=True)
```

```python
data["n"], data["name"] = data[0].str.split(':', 1).str
data
```

```python
data["o"], data["occupation"]= data[1].str.split(':', 1).str
data
```

```python
data.drop([0,1,"n","o"], axis = 1, inplace = True)
```

```python
data["occupation"] = data["occupation"].str.replace("}","").str.replace("]","").str.replace(";",",")
```

```python
data
```

```python
df["Speaker_Name"], df["Speaker_Occupation"] = data["name"], data["occupation"]
```

```python
df
```

```python
dat = df['Topics'].str.split(',', 9, expand=True)
dat
```

```python
dat.drop([0,2,4,6,8], axis = 1, inplace=True)
```

```python
dat
```

```python
# dat.to_csv("Topics.csv", index=False)
dat["n"], dat["name"] = dat[1].str.split(':', 1).str
dat["n0"], dat["name0"] = dat[3].str.split(':', 1).str
dat["n1"], dat["name1"] = dat[5].str.split(':', 1).str
dat["n2"], dat["name2"] = dat[7].str.split(':', 1).str
```

```python
dat0 = dat[["name","name0","name1","name2"]]
```

```python
for i in dat0:
    dat0[i] = dat0[i].str.replace("","").str.replace("}","")
    dat0 = pd.DataFrame(dat0)
```

```python
dat0
```

```python
# 

dat0["name"] = dat0["name"].astype("string")
dat0["name0"] = dat0["name0"].astype("string")
dat0["name1"] = dat0["name1"].astype("string")
dat0["name2"] = dat0["name2"].astype("string")
```

```python
dat0 = dat0.fillna("")
```

```python
dat0.isnull().sum()
```

```python
dat0.info()
```

```python
#df[['First_Name', 'Last_Name']].apply(lambda x: '_'.join(x), axis=1)

dat0["Topics_Names"] = dat0[["name","name0","name1","name2"]].apply(lambda x: ",".join(x), axis=1)
```

```python
df["Topics_Names"] = dat0["Topics_Names"]
```

```python
df
```

```python
df.drop(["Speakers", "Subtitles", "Topics"], axis=1, inplace =True)
```

```python
df.info()
```

```python
df = df[["Id", "Title", "Duration","Likes","Total_Views","Published_Date","Speaker_Name","Speaker_Occupation","Event","Topics_Names"]]
```

```python
df["Topics"] = df["Topics_Names"].str.replace('"', '')
df["Speaker_Title"] = df["Speaker_Occupation"].str.replace('"', '')
df["Speaker_Name"] = df["Speaker_Name"].str.replace('"', '')
```

```python
df.drop(["Speaker_Occupation","Topics_Names"], axis=1, inplace=True)
```

```python
df = df[["Id", "Title", "Duration","Likes","Total_Views","Published_Date","Speaker_Name","Speaker_Title","Event","Topics"]]
```

```python
df["Topics"] = df["Topics"].str.replace(",",", ").str.replace(" ,",",")
```

```python
df
```

```python
df.to_csv("TheTEDTalks.csv", index= False)
```
